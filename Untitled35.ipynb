```python
import requests
url="https://api.fda.gov/drug/event.json?limit=100"
response=requests.get(url)
data=response.json()
```

```python
import pandas as pd
df_uncleaned = pd.json_normalize(data['results'])
```

```python
df_uncleaned
```

```python
df_uncleaned.info()
```

```python
df=df_uncleaned[['serious','patient.patientonsetage','patient.reaction','patient.drug','occurcountry']]
```

```python
df.head()
```

```python
df['patient.reaction'].iloc[1]
```

```python
df['patient.drug'].iloc[1]
```

```python
# Convert patient.reaction into a list with all the reactionmeddrapt
df['reactions_clean'] = df['patient.reaction'].apply(
    lambda lst: [x['reactionmeddrapt'] for x in lst] if isinstance(lst, list) else []
)
```

```python
df.head()
```

```python
df['drug_name']=df['patient.drug'].apply(lambda lst: [x['medicinalproduct'] for x in lst] if isinstance(lst,list) else [])
```

```python
df=df.drop(columns=['patient.reaction','patient.drug'])
```
