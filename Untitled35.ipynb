```python
def clean_data_better(df_uncleaned):
    import pandas as pd

    records = []

    for _, row in df_uncleaned.iterrows():

        report_id = row['safetyreportid']
        serious = row['serious']
        age = row['patient.patientonsetage']
        country = row['occurcountry']

        drugs = row['patient.drug']
        reactions = row['patient.reaction']

        # Extract drug names
        drug_list = []
        if isinstance(drugs, list):
            for d in drugs:
                name = d.get('medicinalproduct')
                if name:
                    drug_list.append(name)

        # Extract reactions
        reaction_list = []
        if isinstance(reactions, list):
            for r in reactions:
                name = r.get('reactionmeddrapt')
                if name:
                    reaction_list.append(name)

        # Skip empty
        if not drug_list or not reaction_list:
            continue

        # Remove duplicates inside a report
        drug_list = list(set(drug_list))
        reaction_list = list(set(reaction_list))

        # Create pairs (controlled)
        for drug in drug_list:
            for reaction in reaction_list:
                records.append({
                    'report_id': report_id,
                    'drug_name': drug,
                    'reactions': reaction,
                    'patient_age': age,
                    'serious': serious,
                    'occurcountry': country
                })

    df = pd.DataFrame(records)

    # Convert types
    df['serious'] = pd.to_numeric(df['serious'], errors='coerce')
    df['patient_age'] = pd.to_numeric(df['patient_age'], errors='coerce')

    # Drop missing
    df = df.dropna(subset=['drug_name', 'reactions'])

    df = df.reset_index(drop=True)

    return df
```

```python
df=clean_data_better(df_raw)
```

```python
#Making comparision
serious=df[df['serious']==1]['reactions'].value_counts(normalize=True).head(10)
unserious=df[df['serious']==2]['reactions'].value_counts(normalize=True).head(10)

comparision=pd.DataFrame({
    'serious':serious,
    'non_serious':unserious
}).fillna(0)

comparision['difference']=comparision['serious']-comparision['non_serious']
comparision.sort_values(by='difference',ascending=False)
```

Certain reactions such as vomiting, pneumonia, and rash show a higher relative frequency in serious cases compared to non-serious ones, suggesting potential clinical severity. In contrast, reactions like nausea, fatigue, and headache are more commonly observed in non-serious cases, indicating they may be less severe or more routine adverse effects.

Positive difference → more associated with serioua

Negative difference → more common in non-serious
